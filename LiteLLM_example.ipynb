### [OpenAI](https://docs.litellm.ai/docs/providers/openai)

## List all the public models from the Litellm public catalog

```python
import litellm
import json

def get_models_for_provider(provider_name: str):
  model_cost_map = litellm.model_cost
  print(model_cost_map.keys())
  provider_models = [
    model for model in model_cost_map.keys()
    if model.startswith(f"{provider_name.lower()}/")
  ]
  return sorted(provider_models)

#print(json.dumps(get_models_for_provider("openrouter"), indent=2))
#print(json.dumps(get_models_for_provider("azure_ai"), indent=2))
#print(json.dumps(get_models_for_provider("openai"), indent=2))
#print(json.dumps(get_models_for_provider("anthropic"), indent=2))
#print(json.dumps(get_models_for_provider("bedrock"), indent=2))
#print(json.dumps(get_models_for_provider("gemini"), indent=2))
```

## Model not found error

```python
from litellm import completion
from litellm import NotFoundError
from google.colab import userdata

try :
    response = completion(
      api_base="https://openhands-test.services.ai.azure.com/",
      api_key=userdata.get('AZURE_API_KEY'),
      model = "azure/dude",
      messages = [{ "content": "Hello, how are you?","role": "user"}]
  )
except NotFoundError as e:
    print(f"Azure OpenAI Error Message: {e.message}")
    print(f"HTTP Status Code: {e.status_code}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
else :
    print(response)
```

## Rate limit error

```python
from litellm import completion
from litellm import RateLimitError
from google.colab import userdata
from litellm.exceptions import RateLimitError

try :
  response = completion(
    api_key=userdata.get('MOONSHOT_API_KEY'),
    api_base="https://api.moonshot.ai/v1",
    model = "moonshot/moonshot-v1-8k",
    messages = [{ "content": "Hello, how are you?","role": "user"}]
  )
except RateLimitError as e:
    print(f"Moonshot API Error Message: {e.message}")
    print(f"HTTP Status Code: {e.status_code}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
else :
    print(response)
```
