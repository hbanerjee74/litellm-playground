### [OpenAI](https://docs.litellm.ai/docs/providers/openai)

```python
from litellm import completion
from google.colab import userdata

response = completion(
    api_base = "https://api.openai.com/v1",
    api_key = userdata.get('OPENAI_API_KEY'),
    model = "gpt-4o",
    messages=[{ "content": "Hello, how are you?","role": "user"}]
)

print("OpenAI Response\n")
print(response)
```

# **LiteLLM Model Discovery Flow**

* At the point of adding model to Studio the user gives API key and model
* The model name can vary - for Azure its the deployment name, for openrouter the admin can set a limited set available.
* We cannot list models since there is no discovery mechanism. We can have a harcoded list of models by provider that we can use to assist; but it should be editable.
* The UI will make a simple LLM call and handle the exception

## Model not found error

```python
from litellm import completion
from google.colab import userdata
from litellm.exceptions import NotFoundError

try :
    response = completion(
      api_base="https://openhands-test.services.ai.azure.com/",
      api_key=userdata.get('AZURE_API_KEY'),
      model = "azure/dude",
      messages = [{ "content": "Hello, how are you?","role": "user"}]
  )
except NotFoundError as e:
    print(f"Azure OpenAI Error Message: {e.message}")
    print(f"HTTP Status Code: {e.status_code}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
else :
    print(response)
```

## Rate limit error

```python
from litellm import completion
from google.colab import userdata
from litellm.exceptions import RateLimitError

try :
  response = completion(
    api_key=userdata.get('MOONSHOT_API_KEY'),
    api_base="https://api.moonshot.ai/v1",
    model = "moonshot/moonshot-v1-8k",
    messages = [{ "content": "Hello, how are you?","role": "user"}]
  )
except RateLimitError as e:
    print(f"Moonshot API Error Message: {e.message}")
    print(f"HTTP Status Code: {e.status_code}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
else :
    print(response)
```
